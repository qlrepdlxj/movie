```python
import glob
import pandas as pd
```

```python
file_list = glob.glob('/home/ubuntu/movie/data/final/*.csv')
rate=[]
name=[]
for lis in file_list:
    data = pd.read_csv(lis)
    sum1 = sum(data['일일관람객'])
    sum2 = sum(data['일일상영횟수'])
    rate.append(sum1/sum2)
    name.append(data['영화명'][0])
total = list(zip(rate,name))
total.sort(reverse=True)
total
```
